```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import random

# Load the CSV file
file_path = '/content/drive/MyDrive/License_Plate_Detector/data/annotations/plateScanner.csv'
df = pd.read_csv(file_path)


# Function to draw bounding box on an image
def draw_bbox(image, xmin, ymin, xmax, ymax):
    cv2.rectangle(image, (xmin, ymin), (xmax, ymax), (255, 0, 0), 2)
    return image

# Select random sample images
sampled_df = df.sample(n=4)

# Plotting the images with bounding boxes
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
axes = axes.flatten()

for i, row in sampled_df.iterrows():
    img_path = row['ImgPath']
    xmin, ymin, xmax, ymax = int(row['Xmin']), int(row['Ymin']), int(row['Xmax']), int(row['Ymax'])
    print(img_path)
    # Read and convert the image
    image = cv2.imread(img_path)
    image= cv2.resize(image, (512, 512))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Draw bounding box
    image = draw_bbox(image, xmin, ymin, xmax, ymax)

    # Plot the image
    axes[i % 4].imshow(image)
    axes[i % 4].set_title(f"Image {i}")
    axes[i % 4].axis('off')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
annot_df = pd.read_csv("/content/drive/MyDrive/License_Plate_Detector/data/annotations/plateScanner.csv")
annot_df.head(15)
```

```python
import ipywidgets as widgets
from IPython.display import display
from PIL import Image as PILImage
import io
import numpy as np
import cv2
import os

# Assume annot_df is a DataFrame already defined in your environment
options = annot_df.ImgPath.unique().tolist()

# Create the dropdown menu
dropdown = widgets.Dropdown(
    options=options,
    value=options[0],
    description='Choose:',
    disabled=False,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='500px')
)

# Create a text box for searching
search_box = widgets.Text(
    placeholder='Type to search...',
    description='Search:',
    disabled=False,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='500px')
)

# Create Next and Previous buttons
next_button = widgets.Button(
    description='Next',
    disabled=False,
    button_style='success',  # 'success' for green button
    tooltip='Click to move to next option',
    icon='arrow-right'
)

prev_button = widgets.Button(
    description='Previous',
    disabled=False,
    button_style='info',  # 'info' for blue button
    tooltip='Click to move to previous option',
    icon='arrow-left'
)

# Function to read and convert image to bytes
def load_image(image_path):
    img = PILImage.open(image_path)
    buf = io.BytesIO()
    img.save(buf, format='PNG')
    return buf.getvalue()

# Function to create image widgets from a list of paths and texts
def create_image_widgets(image_paths, texts):
    image_widgets = []
    for path, text in zip(image_paths, texts):
        image = widgets.Image(
            value=load_image(path),
            format='png',
            layout=widgets.Layout(width='300px', height='150px')
        )
        label = widgets.Label(text)
        image_widgets.append(widgets.VBox([image, label]))
    return image_widgets

# Function to update images based on the selected dropdown value
def update_images(change):
    selected_value = change['new']
    img_df = annot_df[annot_df["ImgPath"] == selected_value]
    os.makedirs("detected_images", exist_ok=True)

    # Clear existing images
    for file in os.listdir("detected_images"):
        os.remove(os.path.join("detected_images", file))
    easyocrs = []
    paddleocrs = []
    confidences = []
    for index, row in img_df.iterrows():
        image = cv2.imread(f'{row.ImgPath}')
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        cropped_image = image[int(row['Ymin']):int(row['Ymax']), int(row['Xmin']):int(row['Xmax'])]
        img = PILImage.fromarray(cropped_image, 'RGB')
        img.save(f'detected_images/{index}.png')
        easyocrs.append(row['EasyOcr'])
        paddleocrs.append(row['PaddleOCR'])
        confidences.append(row['Confidence'])

    dynamic_image_paths = [f'detected_images/{x}' for x in os.listdir('detected_images')]
    dynamic_texts = [f"""Confidence: {round(c,2)}  \n Easy OCR: {e} \n  Paddle OCR: {p}""" for e,p,c in zip(easyocrs, paddleocrs, confidences)]  # Replace with your dynamic texts

    # Update dynamic images
    dynamic_images = create_image_widgets(dynamic_image_paths, dynamic_texts)

    # Update the main image and the dynamic images
    image1.value = load_image(f'{selected_value}')
    image_section.children = [
        image1,
        widgets.VBox(dynamic_images)
    ]

# Function to update dropdown options based on search query
def filter_options(change):
    search_query = change['new'].lower()
    filtered_options = [option for option in options if search_query in option.lower()]
    dropdown.options = filtered_options if filtered_options else ['No match found']
    if filtered_options:
        dropdown.value = filtered_options[0]

# Function to move to the next dropdown option
def next_option(b):
    filtered_options = dropdown.options
    if 'No match found' not in filtered_options:
        current_index = filtered_options.index(dropdown.value)
        next_index = (current_index + 1) % len(filtered_options)
        dropdown.value = filtered_options[next_index]

# Function to move to the previous dropdown option
def prev_option(b):
    filtered_options = dropdown.options
    if 'No match found' not in filtered_options:
        current_index = filtered_options.index(dropdown.value)
        prev_index = (current_index - 1) % len(filtered_options)
        dropdown.value = filtered_options[prev_index]

# Attach the functions to the search box and buttons
search_box.observe(filter_options, names='value')
dropdown.observe(update_images, names='value')
next_button.on_click(next_option)
prev_button.on_click(prev_option)

# Create initial dynamic images (empty for initialization)
dynamic_images = create_image_widgets([], [])

# Create Image placeholders with initial images
image1 = widgets.Image(
    value=load_image(f'{dropdown.value}'),
    format='png',
    layout=widgets.Layout(width='500px', height='300px', border='solid 2px #FF5733', padding='10px')
)

# Organize the layout using VBox and HBox
search_and_dropdown = widgets.VBox([search_box, dropdown])
buttons = widgets.HBox([prev_button, next_button])

# Section for the single image on the left and dynamic images on the right
image_section = widgets.HBox([
    image1,
    widgets.VBox(dynamic_images)
])

# Display the search box, dropdown menu, buttons, and images in a VBox for better organization
ui = widgets.VBox([
    widgets.HTML('<h2>License Plate Detector & OCR</h2>'),
    search_and_dropdown,
    buttons,
    image_section
])

display(ui)
```

```python
from transformers import LlavaNextProcessor, LlavaNextForConditionalGeneration
import torch
from PIL import Image
import requests

processor = LlavaNextProcessor.from_pretrained("llava-hf/llava-v1.6-mistral-7b-hf")

model = LlavaNextForConditionalGeneration.from_pretrained("llava-hf/llava-v1.6-mistral-7b-hf", torch_dtype=torch.float16)
model.to("cuda:0")
```

```python
# prepare image and text prompt, using the appropriate prompt template
url = "/content/4f3fd4b8-5973-4fae-8103-a84ff0863695.png"
# image = Image.open(requests.get(url, stream=True).raw)
image = Image.open(url)
prompt = "[INST] <image>\nplease read the text from this image. the image text is in english. Please make 2 section one for English Letter, 2nd section for English numbers[/INST]"

inputs = processor(prompt, image, return_tensors="pt").to("cuda:0")

# autoregressively complete prompt
output = model.generate(**inputs, max_new_tokens=1000)

print(processor.decode(output[0], skip_special_tokens=True))
```

```python
# test.py
import torch
from PIL import Image
from transformers import AutoModel, AutoTokenizer

model = AutoModel.from_pretrained('openbmb/MiniCPM-Llama3-V-2_5', trust_remote_code=True, torch_dtype=torch.float16)
model = model.to(device='cuda')

tokenizer = AutoTokenizer.from_pretrained('openbmb/MiniCPM-Llama3-V-2_5', trust_remote_code=True)
model.eval()
```

```python
import pandas as pd
import os

def arrange_alpha_digits(text):
  # Remove square brackets
  text = text.replace('[', '').replace(']', '')

  # Split the string into alphabets and digits
  alphabets = []
  digits = []
  for char in text:
    if char.isalpha():
      alphabets.append(char)
    elif char.isdigit():
      digits.append(char)

  # Join the alphabets and digits in the desired order
  result = ''.join(alphabets) + ' ' + ''.join(digits)

  return result

tl_results = []
for image_path in os.listdir("/content/license_plates/validateImages"):
  image = Image.open(f'/content/license_plates/validateImages/{image_path}').convert('RGB')
  question = 'please read the text from this image. Please make sure only return 3 English Alphabets and 4 digits. Please make English alphabets are 3 in numbers and 4 numeric digits "total letters never be more than 7". I need output like in format [ABC 2323]. please do not add square brackets in output'
  msgs = [{'role': 'user', 'content': question}]

  res = model.chat(
      image=image,
      msgs=msgs,
      tokenizer=tokenizer,
      sampling=True, # if sampling=False, beam_search will be used by default
      temperature=0.7,
      # system_prompt='' # pass system_prompt if needed
  )
  print(res)

  ## if you want to use streaming, please make sure sampling=True and stream=True
  ## the model.chat will return a generator
  res = model.chat(
      image=image,
      msgs=msgs,
      tokenizer=tokenizer,
      sampling=True,
      temperature=0.7,
      stream=True
  )

  generated_text = ""
  for new_text in res:
      generated_text += new_text
  tl_results.append({"Image":image_path, "llama3-OCR":arrange_alpha_digits(generated_text)})
true_values_df = pd.read_csv("/content/license_plates/validation_images_data.csv")
ocr_values_df = pd.DataFrame(tl_results)
total_df = true_values_df.merge(ocr_values_df, on='Image',how='inner')
total_df
```

```python
from difflib import SequenceMatcher
# Define a function to calculate the similarity ratio between two strings
def calculate_similarity(plate, ocr_result):
    return SequenceMatcher(None, plate, ocr_result).ratio()

# Calculate the similarity ratio for each row
total_df['Accuracy'] = total_df.apply(lambda row: calculate_similarity(row['LicensePlateNumber'], row['llama3-OCR']), axis=1)
total_df
```
